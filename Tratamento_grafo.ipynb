```python
import csv

arquivo_entrada = "soc-sign-bitcoinalpha.csv"
arquivo_saida = "grafo_sem_tempo.csv"

arestas_vistas = set()
arestas_paralelas = 0
total_linhas = 0

with open(arquivo_entrada, mode='r', encoding='utf-8') as f_in, \
     open(arquivo_saida, mode='w', encoding='utf-8', newline='') as f_out:
    
    leitor = csv.reader(f_in)
    escritor = csv.writer(f_out)
    
    escritor.writerow(["source", "target", "weight"])
    
    for linha in leitor:
        if not linha:  
            continue
            
        origem = linha[0]
        destino = linha[1]
        peso = linha[2]
        
        par = (origem, destino)
        
        if par in arestas_vistas:
            arestas_paralelas += 1
        else:
            arestas_vistas.add(par)
            escritor.writerow([origem, destino, peso])
            
        total_linhas += 1

# Exibe o relatório final no terminal
print("--- RELATÓRIO DO GRAFO ---")
print(f"Total de arestas lidas: {total_linhas}")
print(f"Arestas paralelas identificadas: {arestas_paralelas}")
print(f"Total de arestas únicas salvas: {len(arestas_vistas)}")
print(f"Arquivo gerado: {arquivo_saida}")
```
